```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/jlh/global-health-map/data/separated_gender.csv")
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
X = df.drop(columns=['Location', 'Period', 'Gender'])
y = df["Gender"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
label_encoder.fit(y_test)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
from keras.utils import to_categorical

# Step 2: One-hot encoding
hot_y_train = to_categorical(encoded_y_train)
hot_y_test = to_categorical(encoded_y_test)
hot_y_train
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
hot_y_train.shape
```

```python
X_train.shape
```

```python
from tensorflow.keras.models import Sequential
model = Sequential()
```

```python
from tensorflow.keras.layers import Dense
model.add(Dense(units=100, activation='relu', input_dim=7))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(
    X_train_scaled,
    hot_y_train,
    epochs=250,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, hot_y_test, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model.predict_classes(X_test_scaled[:10])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:10])}")
```
