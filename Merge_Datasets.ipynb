```python
#import our dependencies all at once
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy import stats
import matplotlib.pyplot as plt
import 
```

```python
#read our various csvs
df = pd.read_csv("/Users/jlh/global-health-map/data/lifeExpectancyAtBirth.csv")
dfHale = pd.read_csv("/Users/jlh/global-health-map/data/HALElifeExpectancyAtBirth.csv")
df_30_70 = pd.read_csv("/Users/jlh/global-health-map/data/30-70cancerChdEtc.csv")
```

```python
#begin the fun process of merging our csvs. we'll merge using multiple columns to avoid duplicate entries/fat csv.
df_combined1 = pd.merge(df, dfHale, how='left', left_on=['Location', 'Period', 'Gender'], right_on = ['Location', 'Period', 'Gender'])
df_combined1.head()
```

```python
#continue adding each csv to the pile. 
df_combined2 = pd.merge(df_combined1, df_30_70, how='right', right_on=['Location', 'Period', 'Gender'], left_on = ['Location', 'Period', 'Gender'])
df_combined2.head()
```

```python
#reading our csv
df_suicide = pd.read_csv("/Users/jlh/global-health-map/data/crudeSuicideRates.csv")
```

```python
#notice we're using 'how='right''. because our df_suicide csv has more entries, we're adding it on there. either way works, i prefer to eliminate entries later using .dropna()
df_combined3 = pd.merge(df_combined2, df_suicide, how='right', right_on=['Location', 'Period', 'Gender'], left_on = ['Location', 'Period', 'Gender'])
df_combined3
```

```python
df_alcohol = pd.read_csv("/Users/jlh/global-health-map/data/alcoholSubstanceAbuse.csv")
df_tobacco = pd.read_csv("/Users/jlh/global-health-map/data/tobaccoAge15.csv")
df_poisoning = pd.read_csv("/Users/jlh/global-health-map/data/mortalityRatePoisoning.csv")
df_tobacco["Period"].value_counts()
df_poisoning["Period"].value_counts()
```

```python
df_combined4 = pd.merge(df_combined3, df_alcohol, how='right', right_on=['Location', 'Period', 'Gender'], left_on = ['Location', 'Period', 'Gender'])
df_combined5 = pd.merge(df_combined4, df_poisoning, how='right', right_on=['Location', 'Period', 'Gender'], left_on = ['Location', 'Period', 'Gender'])
df_combined6 = pd.merge(df_combined5, df_tobacco, how='right', right_on=['Location', 'Period', 'Gender'], left_on = ['Location', 'Period', 'Gender'])
```

```python
df_combined6
```

```python
df_combined6.to_csv("data/combined_raw.csv")
```

```python
df_combined_clean = df_combined6.dropna()
```

```python
df_combined_clean.to_csv("data/combined_clean.csv")
df_combined_clean
```

```python
#create a few different final versions, one including both genders while the other summarizes each one. we'll use the gendered
#csv to see if machine learning can predict M/F based on the other categories. As for the genderless df, we'll continue to parse
#and add categories. At some point we'll look for another machine learning algorithm. 
df_gender_separated = df_combined_clean.loc[df_combined_clean["Gender"] != "Both sexes"]
df_all_gender = df_combined_clean.loc[df_combined_clean["Gender"] == "Both sexes"]
df_all_gender
df_gender_separated
```
